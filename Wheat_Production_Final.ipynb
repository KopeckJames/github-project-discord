```python
#Load Prereq Libraries
import os
import sys
import time
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import path
from pathlib import Path
```

```python
wheat_df = pd.read_csv('Resources/world food production.csv')

wheat_df.head()
```

```python
world_wheat = wheat_df.rename(columns={'Entity': 'Country'})

world_wheat.head()
```

```python
world_wheat.info()
```

```python
world_wheat.columns


```

```python
columns_to_drop = ['Maize Production (tonnes)', 'Yams Production (tonnes)', 'Tomatoes Production (tonnes)', 'Tea  Production ( tonnes )', 'Sweet potatoes  Production (tonnes)',
                   'Sunflower seed  Production (tonnes)', 'Sugar cane Production (tonnes)', 'Rice Production ( tonnes)', 'Yams Production (tonnes)',
                   'Soybeans  Production (tonnes)', 'Rye  Production (tonnes)',
                   'Potatoes  Production (tonnes)', 'Oranges  Production (tonnes)',
                   'Peas, dry Production ( tonnes)', 'Palm oil  Production (tonnes)',
                   'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)',
                   'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)',
                   'Bananas  Production ( tonnes)', 'Avocados Production (tonnes)',
                   'Apples Production (tonnes)']

columns_to_drop_existing = [col for col in columns_to_drop if col in world_wheat.columns]

world_wheat.drop(columns=columns_to_drop_existing, axis=1, inplace=True)

world_wheat.head()
```

```python
#For Rice Production, Yams Production and Wheat Production, we're converting the values to billion tonnes
world_wheat['Wheat Production (tonnes)'] = world_wheat['Wheat Production (tonnes)'] / 1000000000


world_wheat.head()
```

```python
world_wheat.rename(columns={'Wheat Production (tonnes)': 'Wheat Production (billion tonnes)'}, inplace=True)

world_wheat.head()
```

```python
world_wheat = world_wheat[['Country', 'Year', 'Wheat Production (billion tonnes)']]
world_wheat.head()
```

```python
#Get Dataframe for all proudctions in each year
yearly_production = world_wheat.groupby('Year')['Wheat Production (billion tonnes)'].sum().reset_index()
yearly_production.tail()

```

```python
#Graph of Russia Wheat Production over the years
russia_wheat = world_wheat[world_wheat['Country'] == 'Russia']
russia_wheat.groupby('Year')['Wheat Production (billion tonnes)'].sum().plot(kind='bar', figsize=(12, 6))
plt.title('Russia Wheat Production (billion tonnes) over the years')
plt.ylabel('Wheat Production (billion tonnes)')
xticks = np.arange(1961, 2022, 1)
plt.show()
```

```python
#Graph of yearly wheat production worldwide (line graph)
yearly_production.plot(x='Year', y='Wheat Production (billion tonnes)', kind='line', figsize=(12, 6))
plt.title('Worldwide Wheat Production (billion tonnes) over the years')
plt.ylabel('Wheat Production (billion tonnes)')
plt.xticks(yearly_production['Year'], rotation=90)
#plt.savefig('Resources/World_Wheat_Production_Line.png')
plt.show()
```

```python
#Bar graph of yearly wheat production worldwide
yearly_production.plot(x='Year', y='Wheat Production (billion tonnes)', kind='bar', figsize=(12, 6))
plt.title('Worldwide Wheat Production (billion tonnes) over the years')
plt.ylabel('Wheat Production (billion tonnes)')
#plt.savefig('Resources/World_Wheat_Production_Bar.png')
plt.show()
```
