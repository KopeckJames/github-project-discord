```python
#Load Prereq Libraries
import os
import sys
import time
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import path
from pathlib import Path
```

```python
world_wheat = pd.read_csv('Resources/wheat-yields.csv')

world_wheat.head()
```

```python
world_wheat = world_wheat.rename(columns={'Wheat yield': 'Wheat Yield (tonnes)'})
world_wheat = world_wheat.rename(columns={'Entity': 'Country'})

world_wheat.head()
```

```python
#Gain Total Wheat Production by Year per Country
total_wheat_yield_per_year = world_wheat.groupby('Year')['Wheat Yield (tonnes)'].sum().reset_index()

total_wheat_yield_per_year.tail()
```

```python
#DF for Total Wheat Production by US
us_wheat = world_wheat[world_wheat['Country'] == 'United States']
us_wheat.head()
```

```python
#Make Bar Graph for World Wheat Production
plt.figure(figsize=(10, 6))
plt.bar(total_wheat_yield_per_year['Year'], total_wheat_yield_per_year['Wheat Yield (tonnes)'])
plt.xlabel('Year')
plt.ylabel('Wheat Yield (tonnes)')
plt.title('World Wheat Production by Year')
plt.show()
```

```python
#Only show data for Year values 1960-2022
total_wheat_yield_per_year = total_wheat_yield_per_year[total_wheat_yield_per_year['Year'] >= 1960]
total_wheat_yield_per_year.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(total_wheat_yield_per_year['Year'], total_wheat_yield_per_year['Wheat Yield (tonnes)'])
plt.xlabel('Year')
plt.ylabel('Wheat Yield (tonnes)')
plt.title('World Wheat Production by Year')
plt.show()
```

```python
#Line Graph for world wheat production
plt.figure(figsize=(10, 6))
plt.plot(total_wheat_yield_per_year['Year'], total_wheat_yield_per_year['Wheat Yield (tonnes)'])
plt.xlabel('Year')
plt.ylabel('Wheat Yield (tonnes)')
plt.title('World Wheat Production by Year')
plt.show()
```
